```python
import tensorflow as tf

# 1. MNIST 데이터셋 임포트 - 손글씨
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
# 2. 데이터 전처리
x_train, x_test = x_train/255.0, x_test/255.0 # 흰, 검 - 흑백 영상 표준화 (gray scale)
```

```python
# 3. 모델 구성
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(512, activation=tf.nn.relu),
    tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])
```

```python
# 4. 모델 컴파일
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# 5. 모델 훈련
model.fit(x_train, y_train, epochs=5)
```

```python
# 6. 정확도 평가
test_loss, test_acc = model.evaluate(x_test, y_test)
print('테스트 정확도:', test_acc)
```

# Shapes

```python
import os
import cv2
import numpy as np

# 데이터셋 폴더 경로와 클래스(레이블) 목록 설정
dataset_folder = '/content/drive/MyDrive/Colab Notebooks/datasets/shapes'
class_labels = ['circles', 'squares', 'triangles']

# 이미지와 레이블을 저장할 리스트 생성
images = []
labels = []

# 클래스(레이블)별로 순회하며 이미지와 레이블 수집
for i, class_label in enumerate(class_labels):
    folder_path = os.path.join(dataset_folder, class_label)
    image_files = os.listdir(folder_path)

    for image_file in image_files:
        image_path = os.path.join(folder_path, image_file)
        image = cv2.imread(image_path)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # 이미지 색상 채널 변환
        image = image / 255.0  # 이미지 정규화
        images.append(image)
        labels.append(i)  # 클래스(레이블) 인덱스 추가
```

```python
# 리스트를 NumPy 배열로 변환
x_data = np.array(images)
y_data = np.array(labels)

# 데이터 확인
print('데이터 shape:', x_data.shape)  # (이미지 개수, 이미지 너비, 이미지 높이, 채널 수)
print('레이블 shape:', y_data.shape)  # (이미지 개수,)
```

```python
from sklearn.model_selection import train_test_split

# 데이터셋을 훈련 데이터와 테스트 데이터로 나눔
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=42)

# 데이터 분할 결과 확인
print('훈련 데이터 shape:', x_train.shape)
print('훈련 레이블 shape:', y_train.shape)
print('테스트 데이터 shape:', x_test.shape)
print('테스트 레이블 shape:', y_test.shape)
```

## Shapes - ANN

```python
import tensorflow as tf

# 2. 데이터 전처리
# x_train, x_test = x_train/255.0, x_test/255.0

# 3. 모델 구성
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28, 3)),
    tf.keras.layers.Dense(512, activation=tf.nn.relu),
    tf.keras.layers.Dense(256, activation=tf.nn.relu),
    tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])

# 4. 모델 컴파일
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
import matplotlib.pyplot as plt

# 모델 훈련
history = model.fit(x_train, y_train, epochs=100, batch_size=32, validation_data=(x_test, y_test))
```

```python
# Loss
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Accuracy
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
# 6. 정확도 평가
test_loss, test_acc = model.evaluate(x_test, y_test)
print('테스트 정확도:', test_acc)
```

## Shapes - CNN
비슷한 이미지인데 위치가 달라지면 NN 말고 CNN이 적합함

```python
import os
import numpy as np
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
```

```python
# 데이터 로드 및 전처리
data_dir = '/content/drive/MyDrive/Colab Notebooks/datasets/shapes'
batch_size = 32
img_height = 180
img_width = 180
```

```python
train_datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)
train_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='training')
```

```python
validation_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation')
```

```python
model = Sequential([
    # 첫 번째 컨볼루션 블록
    Conv2D(32, (3, 3), activation='relu', input_shape=(img_height, img_width, 3)),
    MaxPooling2D(2, 2),

    # 두 번째 컨볼루션 블록
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Dropout(0.25),

    # 세 번째 컨볼루션 블록
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),

    # 네 번째 컨볼루션 블록
    Conv2D(256, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Dropout(0.25),

    # 네 번째 컨볼루션 블록
    Conv2D(512, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),

    # Flatten 및 완전 연결 레이어
    Flatten(),
    Dense(512, activation='relu'),
    Dense(3, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# 모델 훈련 및 평가
history = model.fit(train_generator, epochs=50, batch_size=48, validation_data=validation_generator)
```

```python
# 그래프를 그리기 위한 matplotlib 라이브러리
import matplotlib.pyplot as plt
# Training Loss VS Validation Loss 비교를 위한 그래프 그리기
plt.plot(history.history['loss'], 'y', label='train loss')
plt.plot(history.history['val_loss'], 'r', label='val loss')
# y축 범위 설정
plt.ylim([0.0, 2.5])
# 각 축의 이름 정하기
plt.xlabel('epoch')
plt.ylabel('loss')
# 각 그래프의 설명 위치 설정 후 표시
plt.legend(loc='upper left')
plt.show()
```

```python
# Training Accuracy VS Validation Accuracy 비교를 위한 그래프 그리기
plt.figure(figsize=(6,8))
plt.plot(history.history['accuracy'], 'b', label='train acc')
plt.plot(history.history['val_accuracy'], 'g', label='val acc')
# y축 범위 설정
plt.ylim([0.2, 1.0])
# 각 축의 이름 정하기
plt.xlabel('epoch')
plt.ylabel('accuray')
# 각 그래프의 설명 위치 설정 후 표시
plt.legend(loc='upper left')
# 그래프 그리기
plt.show()
```
