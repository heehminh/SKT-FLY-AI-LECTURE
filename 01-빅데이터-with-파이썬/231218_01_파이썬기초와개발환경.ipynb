# 웹에서 이미지 불러오기

```python
import requests
from PIL import Image
from io import BytesIO
```

```python
response = requests.get('https://www.naver.com')
print(response.text)
```

```python
# 인터넷에서 이미지 가져오기
response = requests.get("https://search.pstatic.net/common/?src=http%3A%2F%2Fblogfiles.naver.net%2FMjAyMzExMjlfMTky%2FMDAxNzAxMjMxNTA3NTQ0.TMBl7yxz5R7xvDrjyMZs5_CN04135Y08YEGYl3YWKY0g.Ha91B4S7m8m31B-S6kzP_9i67lPfuRXdB1A9WK4mKU8g.JPEG.a3574284%2F3472511939180767549_20231123234200265.jpg&type=sc960_832")
```

```python
response.content # x로 시작: 16진수 (바이너리 데이터를 보낼 땐 안깨지게 16진수 = 2^4로 보냄)
```

```python
BytesIO(response.content) # 실제 이미지의 바이트코드로 복원
```

```python
image = Image.open(BytesIO(response.content)) # 이미지로 복원
image
```
