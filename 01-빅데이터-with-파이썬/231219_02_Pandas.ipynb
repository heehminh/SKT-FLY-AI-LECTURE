# 결측치

````python
import numpy as np
import pandas as pd

z = [['Jackson', 68, True, np.NaN, 'A'],
  ['Liam', 74, True, 1.1, 'A'],
  ['Emma', np.NaN, False, 0.89, 'C'],
  ['James', 78, np.NaN, 0.91, 'C'],
  ['Ava', 54, False, 0.73, 'C'],
  ['Henry', 69, True, np.NaN, 'C']]

df = pd.DataFrame(data= z, columns=['Name', 'Weight', 'Option', 'Rate', 'Type'])
df
````

````python
# 결측치 확인
df.isna()
````

````python
# 결측치 개수 확인
df.isna().sum()
````

````python
# 결측치 포함행 삭제
df.dropna()
````

````python
# 결측치를 0으로 변경
df.fillna(0)
````

````python
# 특정 컬럼의 결측치만 설정  ('Weight' 컬럼의 결측치를 해당 컬럼의 평균으로 변경)
m = df['Weight'].mean()
print('mean :', m)
df['Weight'].fillna(m, inplace=True)
df
````

````python
# 원본 df는 변경하지 않고 결측치가 삭제된 결과는 re 변수에 저장
re = df.dropna()
df
````

````python
# 원본데이터 프레임 수정
re = df.dropna(inplace=True)
df
````

# 인덱싱과 슬라이싱

````python
z = [['Jackson', 68, True, np.NaN, 'A'],
    ['Liam', 74, True, 1.1, 'A'],
    ['Emma', np.NaN, False, 0.89, 'C'],
    ['James', 78, np.NaN, 0.91, 'C'],
    ['Ava', 54, False, 0.73, 'C'],
    ['Henry', 69, True, np.NaN, 'C']]

df = pd.DataFrame(data= z, columns=['Name', 'Weight', 'Option', 'Rate', 'Type'])
df
````

````python
df.loc[2:4] # 2행 - 4행까지 모든 컬럼의 데이터 추출
````

````python
df.loc[2:4, ['Name', 'Weight']] # 2행 - 4행까지 모든 컬럼의 데이터 추출
````

````python
df.loc[2:4, 'Weight'] # 하나의 컬럼만 추출
````

````python
# 조건에 맞는 행만 추출 (Rate 컬럼의 값이 0.8보다 큰 행의 Weight, Rate, Type만 추출)
x = df.loc[df['Rate']>0.8, ['Weight', 'Rate', 'Type']]
print('type : ', type(x)) # DataFrame
x
````

````python
# Option 컬럼이 True인 행의 모든 열 추출
x = df.loc[df['Option']==True]
x
````

````python
# 다중 조건 사용하기
x = df.loc[(df['Rate']>0.8) &(df['Option']==True), :]
x
````

````python
# 행과 열을 숫자로 지정: 추출한 컬럼을 숫자로 지정
x = df.iloc[2:4, 1:-1]
x
````

````python
# 특정 데이터 타입을 가지는 컬럼만 추출
x = [['Jackson', 68, True, 0.82, 'A'],
    ['Liam', 74, True, 1.1, 'B'],
    ['Emma', 52, False, 0.89, 'C'],
    ['James', 78, True, 0.91, 'C'],
    ['Ava', 54, False, 0.73, 'B'],
    ['Henry', 69, True, 0.78, 'C']]

df = pd.DataFrame(x, columns=['Name', 'Weight', 'Option', 'Rate', 'Type'])
df.info()
````
