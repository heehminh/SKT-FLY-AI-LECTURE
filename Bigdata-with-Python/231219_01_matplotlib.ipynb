## 폰트 설치 후 런터임 재시작

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')

x = np.arange(0, 10)
y = x**2
plt.plot(x, y)
plt.title('플롯 그래프')
plt.xlabel('X축')
plt.ylabel('Y축')
plt.show()
```

```python
# 색상과 스타일 지정
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.sin(x + 0.5)
y3 = np.sin(x + 1)

plt.plot(x, y1, 'r--') # color, 선의 종류, 마커
plt.plot(x, y2, 'b.')
plt.plot(x, y3, 'g-.')

plt.title('sin 함수')
plt.xlabel('X축')
plt.ylabel('Y축')
plt.show()
```

```python
# 범례 표시
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.sin(x + 0.5)
y3 = np.sin(x + 1)

plt.plot(x, y1, 'r--', label='Sin A')
plt.plot(x, y2, 'b.', label='Sin B')
plt.plot(x, y3, 'g-.', label='Sin C')
plt.legend()

plt.title('sin 함수')
plt.xlabel('X축')
plt.ylabel('Y축')

plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.sin(x + 0.5)
y3 = np.sin(x + 1)

plt.plot(x, y1, 'r--', label='Sin A')
plt.plot(x, y2, 'b.', label='Sin B')
plt.plot(x, y3, 'g-.', label='Sin C')
plt.legend(loc='upper right') # 범례의 위치

plt.title('sin 함수')
plt.xlabel('X축')
plt.ylabel('Y축')

plt.show()
```

```python
# 여러개의 그래프 그리기
x = np.linspace(0, 10, 30)
y = np.sin(x)
z = np.cos(x)

plt.style.use('ggplot')
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.plot(x, y, 'r-')

plt.subplot(2, 2, 2)
plt.plot(x, z, 'g--')

plt.subplot(2, 2, 3)
plt.plot(x, z, 'b-.')

plt.show()
```

# 산점도: 분석 단계에서 사용

```python
x = np.random.rand(30)
y = np.random.rand(30)
plt.scatter(x,y)
plt.show()
```

```python
# 크기와 색상
n= 50
x = np.random.rand(n)
y = np.random.rand(n)
area = (30 * np.random.rand(n))**2
colors = np.random.rand(n)
plt.scatter(x, y, s=area, c=colors, alpha=0.7)
plt.show()
```

# 막대그래프

## bar()

```python
x = ['A','B','C','D']
y = [5. , 25. , 50. , 20.]
plt.bar(x, y)
plt.title('카테고리별 수치 비교를 나타내는 막대그래프')
plt.xlabel('X축')
plt.ylabel('Y축')
plt.show()
```

## barh()

```python
x = ['A','B','C','D']
y = [5. , 25. , 50. , 20.]
plt.barh(x, y)

plt.title("X축과 Y축을 바꾼 가로 막대 그래프")
plt.xlabel('Y축')
plt.ylabel('X축')


plt.show()
```

## bar() 색상 변경

```python
name = ['A','B','C','D']
y = [5. , 25. , 50. , 20.]
z = [8. , 20. , 25. , 6.]
fig, ax = plt.subplots(1, 2)
fig.set_size_inches(12, 5)
ax[0].bar(name, y, color='#33aaff')
ax[1].bar(name, z, color=['r', 'g', 'b', 'k'])
plt.show()
```

# Boxplot

```python
x = np.random.normal(10, 3, 100)
y = np.random.normal(13, 3, 100)
z = np.random.normal(8, 5, 100)
plt.boxplot([x, y, z])
plt.xticks([1, 2, 3], ['A', 'B', 'C'])
plt.show()
```

```python
# 박스 내부의 색상과 데이터를 표시하는 마커의 색상을 설정
x = np.random.normal(10, 3, 100)
y = np.random.normal(13, 3, 100)
z = np.random.normal(8, 5, 100)
plt.boxplot([x, y, z], patch_artist=True, sym='ro')
plt.xticks([1, 2, 3], ['A', 'B', 'C'])
plt.show()
```
